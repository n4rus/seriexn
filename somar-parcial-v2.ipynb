```python
lista=[]
serie=numpy.float128(tamanho)
xni=numpy.float128(indice)

listaf=[]

while xni<=serie:
    #soma parcial da sequencia modificada
    if xni <= 1:
        partx=1/xni
        #print para testes
        #print(partx)
        lista.append(partx)
        xni=xni+1
    if xni > 1:
        partx=partx+1/xni
        xni=xni+1
        lista.append(partx)
print(lista[900000:1000000])
```
